```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation

# Simulation parameters
GRID_SIZE = 50           # Size of the grid (50x50)
NUM_BOTS = 10            # Number of microbots
NUM_RESOURCES = 20       # Number of resources
MAX_STEPS = 500          # Maximum simulation steps
BASE_LOCATION = (25, 25) # Base location at the center of the grid
DETECTION_RANGE = 3      # Detection range for resources

# Initialize the grid
grid = np.zeros((GRID_SIZE, GRID_SIZE))

# Place resources randomly on the grid
np.random.seed(0)  # For reproducibility
resource_positions = set()
while len(resource_positions) < NUM_RESOURCES:
    pos = (np.random.randint(0, GRID_SIZE), np.random.randint(0, GRID_SIZE))
    if pos != BASE_LOCATION:
        resource_positions.add(pos)
        grid[pos] = 2  # Mark resource cells with 2

# Initialize microbots at random positions
bot_positions = []
for _ in range(NUM_BOTS):
    while True:
        pos = (np.random.randint(0, GRID_SIZE), np.random.randint(0, GRID_SIZE))
        if grid[pos] == 0 and pos != BASE_LOCATION:
            bot_positions.append({
                'position': pos,
                'has_resource': False
            })
            grid[pos] = 1  # Mark bot positions with 1
            break

# Set up the plot
fig, ax = plt.subplots()
im = ax.imshow(grid, cmap='terrain', origin='lower')
ax.plot(BASE_LOCATION[1], BASE_LOCATION[0], 'ro', markersize=5, label='Base')
ax.set_title('Microbots Simulation')
ax.legend(loc='upper right')

def update(frame_num):
    global bot_positions, resource_positions, grid

    new_grid = np.zeros((GRID_SIZE, GRID_SIZE))
    new_grid[BASE_LOCATION] = 3  # Base location

    for bot in bot_positions:
        x, y = bot['position']

        if bot['has_resource']:
            # Move towards the base
            dx = np.sign(BASE_LOCATION[0] - x)
            dy = np.sign(BASE_LOCATION[1] - y)
        else:
            # Detect resources within detection range
            detected = False
            for dx in range(-DETECTION_RANGE, DETECTION_RANGE + 1):
                for dy in range(-DETECTION_RANGE, DETECTION_RANGE + 1):
                    nx, ny = x + dx, y + dy
                    if 0 <= nx < GRID_SIZE and 0 <= ny < GRID_SIZE:
                        if (nx, ny) in resource_positions:
                            # Move towards the resource
                            dx = np.sign(nx - x)
                            dy = np.sign(ny - y)
                            detected = True
                            break
                if detected:
                    break
            if not detected:
                # Random movement
                dx, dy = np.random.choice([-1, 0, 1], 2)

        # Update bot position
        new_x = (x + dx) % GRID_SIZE
        new_y = (y + dy) % GRID_SIZE

        # Check for resource pickup
        if not bot['has_resource'] and (new_x, new_y) in resource_positions:
            bot['has_resource'] = True
            resource_positions.remove((new_x, new_y))

        # Check for resource drop-off
        if bot['has_resource'] and (new_x, new_y) == BASE_LOCATION:
            bot['has_resource'] = False

        bot['position'] = (new_x, new_y)
        new_grid[new_x, new_y] = 1  # Mark bot positions

    # Mark remaining resources
    for pos in resource_positions:
        new_grid[pos] = 2

    im.set_array(new_grid)
    grid = new_grid
    return [im]

# Create animation
ani = animation.FuncAnimation(fig, update, frames=MAX_STEPS, interval=50, blit=True)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation

# Simulation parameters
PIPE_LENGTH = 100       # Length of the pipe
PIPE_WIDTH = 20         # Width of the pipe
NUM_BOTS = 20           # Number of microbots
MAX_STEPS = 200         # Maximum simulation steps
DETECTION_RANGE = 2     # Detection range for the clog
CLOG_POSITION = (50, 10)     # Position of the clog (center)
CLOG_SIZE = (5, 10)     # Size of the clog (length in pipe, width across pipe)
BOT_START_AREA = (0, PIPE_LENGTH // 10)  # Area where bots start

# Initialize the pipe grid
pipe = np.zeros((PIPE_WIDTH, PIPE_LENGTH))

# Place the clog in the pipe
clog_start_x = max(0, CLOG_POSITION[0] - CLOG_SIZE[0] // 2)
clog_end_x = min(PIPE_LENGTH, CLOG_POSITION[0] + CLOG_SIZE[0] // 2)
clog_start_y = max(0, CLOG_POSITION[1] - CLOG_SIZE[1] // 2)
clog_end_y = min(PIPE_WIDTH, CLOG_POSITION[1] + CLOG_SIZE[1] // 2)

pipe[clog_start_y:clog_end_y, clog_start_x:clog_end_x] = 2  # Mark clog cells with 2

# Initialize microbots at the starting area
np.random.seed(0)  # For reproducibility
bot_positions = []
while len(bot_positions) < NUM_BOTS:
    x = np.random.randint(BOT_START_AREA[0], BOT_START_AREA[1])
    y = np.random.randint(0, PIPE_WIDTH)
    if pipe[y, x] == 0:
        bot_positions.append({'position': (y, x)})
        pipe[y, x] = 1  # Mark bot positions with 1

# Set up the plot
fig, ax = plt.subplots(figsize=(12, 6))
im = ax.imshow(pipe, cmap='coolwarm', origin='lower', interpolation='nearest')
ax.set_title('Microbot Swarm Unclogging a Pipe')
ax.set_xlabel('Pipe Length')
ax.set_ylabel('Pipe Width')

def update(frame_num):
    global bot_positions, pipe

    new_pipe = np.copy(pipe)

    for bot in bot_positions:
        y, x = bot['position']

        # Detect clog within detection range
        clog_detected = False
        for dy in range(-DETECTION_RANGE, DETECTION_RANGE + 1):
            for dx in range(-DETECTION_RANGE, DETECTION_RANGE + 1):
                ny, nx = y + dy, x + dx
                if 0 <= ny < PIPE_WIDTH and 0 <= nx < PIPE_LENGTH:
                    if pipe[ny, nx] == 2:
                        # Move towards the clog
                        move_y = np.sign(ny - y)
                        move_x = np.sign(nx - x)
                        clog_detected = True
                        break
            if clog_detected:
                break
        if not clog_detected:
            # Move forward (right) or random movement
            if np.random.rand() < 0.8:
                move_x = 1  # Prefer moving forward
                move_y = 0
            else:
                move_x = np.random.choice([-1, 0, 1])
                move_y = np.random.choice([-1, 0, 1])

        # Update bot position
        new_x = x + move_x
        new_y = y + move_y

        # Keep within bounds
        new_x = max(0, min(PIPE_LENGTH - 1, new_x))
        new_y = max(0, min(PIPE_WIDTH - 1, new_y))

        # Check if the new position is not occupied
        if new_pipe[new_y, new_x] == 0 or new_pipe[new_y, new_x] == 2:
            # Remove bot from old position
            new_pipe[y, x] = 0
            # Interaction with clog
            if new_pipe[new_y, new_x] == 2:
                # Remove part of the clog
                new_pipe[new_y, new_x] = 0
            # Place bot in new position
            new_pipe[new_y, new_x] = 1
            bot['position'] = (new_y, new_x)
        else:
            # Bot stays in place if the position is occupied
            pass

    pipe = new_pipe
    im.set_array(pipe)
    return [im]

# Create animation
ani = animation.FuncAnimation(fig, update, frames=MAX_STEPS, interval=200, blit=True)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation

# Simulation parameters
PIPE_LENGTH = 100       # Length of the pipe
PIPE_WIDTH = 20         # Width of the pipe
NUM_BOTS = 20           # Number of microbots
MAX_STEPS = 500         # Increased simulation steps
DETECTION_RANGE = 5     # Increased detection range
CLOG_POSITION = (80, 10)  # Moved clog further to test bot's ability to reach
CLOG_SIZE = (5, 10)     # Size of the clog (length in pipe, width across pipe)
BOT_START_AREA = (0, 10)  # Bots starting closer to the clog

# Initialize the pipe grid
pipe = np.zeros((PIPE_WIDTH, PIPE_LENGTH))

# Place the clog in the pipe
clog_start_x = max(0, CLOG_POSITION[0] - CLOG_SIZE[0] // 2)
clog_end_x = min(PIPE_LENGTH, CLOG_POSITION[0] + CLOG_SIZE[0] // 2)
clog_start_y = max(0, CLOG_POSITION[1] - CLOG_SIZE[1] // 2)
clog_end_y = min(PIPE_WIDTH, CLOG_POSITION[1] + CLOG_SIZE[1] // 2)

pipe[clog_start_y:clog_end_y, clog_start_x:clog_end_x] = 2  # Mark clog cells with 2

# Initialize microbots at the starting area
np.random.seed(0)  # For reproducibility
bot_positions = []
while len(bot_positions) < NUM_BOTS:
    x = np.random.randint(BOT_START_AREA[0], BOT_START_AREA[1])
    y = np.random.randint(0, PIPE_WIDTH)
    if pipe[y, x] == 0:
        bot_positions.append({'position': (y, x)})
        pipe[y, x] = 1  # Mark bot positions with 1

# Set up the plot
fig, ax = plt.subplots(figsize=(12, 6))
im = ax.imshow(pipe, cmap='coolwarm', origin='lower', interpolation='nearest')
ax.set_title('Microbot Swarm Unclogging a Pipe')
ax.set_xlabel('Pipe Length')
ax.set_ylabel('Pipe Width')

def update(frame_num):
    global bot_positions, pipe

    new_pipe = np.copy(pipe)

    for bot in bot_positions:
        y, x = bot['position']

        # If clog is still present, move towards it
        if np.any(pipe == 2):
            # Compute vector towards the center of the clog
            clog_cells = np.argwhere(pipe == 2)
            clog_center = np.mean(clog_cells, axis=0)
            move_vector = clog_center - np.array([y, x])
            move_y = int(np.sign(move_vector[0]))
            move_x = int(np.sign(move_vector[1]))
        else:
            # Move forward (right) or random movement
            if np.random.rand() < 0.8:
                move_x = 1  # Prefer moving forward
                move_y = 0
            else:
                move_x = np.random.choice([-1, 0, 1])
                move_y = np.random.choice([-1, 0, 1])

        # Update bot position
        new_x = x + move_x
        new_y = y + move_y

        # Keep within bounds
        new_x = max(0, min(PIPE_LENGTH - 1, new_x))
        new_y = max(0, min(PIPE_WIDTH - 1, new_y))

        # Check if the new position is not occupied
        if new_pipe[new_y, new_x] == 0 or new_pipe[new_y, new_x] == 2:
            # Remove bot from old position
            new_pipe[y, x] = 0
            # Interaction with clog
            if new_pipe[new_y, new_x] == 2:
                # Remove part of the clog
                new_pipe[new_y, new_x] = 0
            # Place bot in new position
            new_pipe[new_y, new_x] = 1
            bot['position'] = (new_y, new_x)
        else:
            # Bot stays in place if the position is occupied
            pass

    pipe = new_pipe
    im.set_array(pipe)
    ax.set_title(f'Microbot Swarm Unclogging a Pipe (Step {frame_num})')
    return [im]

# Create animation
ani = animation.FuncAnimation(fig, update, frames=MAX_STEPS, interval=100, blit=True)

plt.show()
```
